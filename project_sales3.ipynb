```python
# rename Unit Price
monthly['Unit_Price']=monthly['Unit Price']
monthly.head()
```

```python
# create a dataframe where EntDate or InvDate is the index
# columns are: Part, Customer, Unit Price - will add promoted yes/no as a binary later
cols = ['EntDate', 'Part','Customer', 'Unit_Price', 'Quantity']
data = monthly[cols]
data.head(20)
```

```python
data.set_index('EntDate', inplace=True)
by_week = data.groupby(['Part']).resample('W')
by_week.Quantity.sum().head(20)
```

```python
df = by_week.Quantity.sum().unstack(1).fillna(0)
```

```python
df.head(20)
```

```python
df.info()
```

```python
df_transposed = df.T
df_transposed.head(20)
```

```python
df_transposed.to_excel('wusthof_time_series.xlsx')
```
