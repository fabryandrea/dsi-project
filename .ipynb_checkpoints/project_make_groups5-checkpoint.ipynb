```python
# huge demand (over 4,000 units)

def identify_huge_demand(dataframe, product_list, n):
    # returns list of products whose demand never exceeds n units
    huge_demand = []
    for product in product_list:
        if dataframe[product].max() > n:
            huge_demand.append(product)
    return huge_demand
```

```python
# turn them back into product groups

# initial data panda df, name: products, columns=datetime, index=product SKUS
# turned into np.ndarray, name: data_arr, same as df without index names and column names



def assign_products_to_clusters(dataframe, dictionary):
    # create a list for all 8 keys
    group_1 = []
    group_2 = []
    group_3 = []
    group_4 = []
    group_5 = []
    group_6 = []
    group_7 = []
    group_8 = []
    for key, value in dictionary.items():
        # use assigned_data.values() as row numbers to find product SKUs in index of dataframe        
        # add corresponding product SKUs to lists
        if key == 0:
            group_1.append(products_df.index[x] for x in list(assigned_data[0]))
        elif key == 1:
            group_2 = [products_df.index[x] for x in list(assigned_data[1])]
        elif key == 2:
            group_3 = [products_df.index[x] for x in list(assigned_data[key])]
        elif key == 3:
            group_4 = [products_df.index[x] for x in list(assigned_data[key])]
        elif key == 4:
            group_5 = [products_df.index[x] for x in list(assigned_data[key])]
        elif key == 5:
            group_6 = [products_df.index[x] for x in list(assigned_data[key])]
        # output: k lists of products 
        return group_1, group_2, group_3, group_4, group_5, group_6
```
