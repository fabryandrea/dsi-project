```python
# rename Unit Price
monthly['Unit_Price']=monthly['Unit Price']
monthly.head()
```

```python
# create a dataframe where EntDate or InvDate is the index
# columns are: Part, Customer, Unit Price - will add promoted yes/no as a binary later
cols = ['EntDate', 'Part','Customer', 'Unit_Price', 'Quantity']
data = monthly[cols]
data.head(20)
```

```python
grouped_data = data.set_index('EntDate').groupby('Part')['Quantity'].resample("M").sum()
grouped_data.head()
```

```python
test=pd.DataFrame(grouped_data)
check=test.reset_index(inplace=True)
test.head()
```

```python
test2 = test.sort_values(['EntDate'])
test2.head()
```

```python
# make EntDate index
test3 = test2.set_index(keys='EntDate')
test3.head()
```
